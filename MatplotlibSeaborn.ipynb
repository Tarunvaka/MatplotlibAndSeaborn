```python
import pandas as pd 

df = pd.read_csv('HistoricalPrices.csv')
df.head()
```

```python
df = df.rename(columns={" Open":"Open", ' High': "High"," Low":"Low"," Close": "Close"})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date')
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
from datetime import datetime
```

```python
plt.plot(df['Date'], df['Close'])
plt.show()
```

```python
plt.plot(df['Date'], df['Open'])
plt.plot(df['Date'],df['Close'])
```

```python
plt.plot(df['Date'], df['Open'], label = 'Open')
plt.plot(df['Date'],df['Close'], label = 'Close')
plt.legend()
plt.show()
```

```python
from calendar import month_name
df['Month'] = pd.Categorical(df['Date'].dt.month_name(), month_name[1:])

df_monthly_mean = df\
    .groupby("Month")\
    .mean()\
    .reset_index()
```

```python
plt.bar(df['Month'],height=df['Close'])
plt.show()
```

```python
plt.scatter(df['Open'], df['Close'])
plt.show()
```

```python
z = np.polyfit(df['Open'], df['Close'],1 )
p = np.poly1d(z)

plt.scatter(df['Open'], df['Close'])
plt.plot(df['Open'], p(df['Open']))
plt.show()
```

```python
plt.plot(df['Date'], df['Open'], color = 'black')
plt.plot(df['Date'], df['Close'], color = 'red')
plt.show()
```

```python
plt.scatter(df['Open'], df['Close'], c = df['Close'], cmap = plt.cm.plasma)
plt.show()
```

```python
plt.scatter(df['Open'], df['Close'])
plt.savefig("DJIA 2024 Scatterplot Open vs Close.png")
```

# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.histplot(data = tips, x = 'total_bill')
```

```python
sns.scatterplot(x = 'total_bill', y='tip', data = tips)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = "sex", size= 'size', sizes = (50,200), data = tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Relationship between total bill and tip")
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri)
```

```python
sns.lineplot(x='timepoint', y ='signal', hue = 'event', style = 'region', markers = True, dashes = False, data = fmri)
plt.xlabel("Timepoint")
plt.ylabel("Signal Intensity")
plt.title("Changes in signal intensity over time")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.barplot(x = 'class', y = 'fare', data = titanic)
```

```python
sns.barplot(x='class', y = 'fare', hue='sex', ci=None, palette="muted", data= titanic)
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Average Fare by Class and Gender on the Titanic")
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.histplot(x = 'petal_length', data = iris)
```

```python
sns.histplot(data=iris, x = 'petal_length', bins = 20, kde=True, color = 'green')
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Distribution of petal lengths in Iris Flowers")
plt.show()
```

```python
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
sns.kdeplot(data=tips, x = 'total_bill', hue = 'time', fill = True, alpha = 0.6, linewidth = 1.5)
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.title("Density Plot of Total Bill by Meal Table")
plt.show()
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue = 'time', data=tips, palette="Set3", linewidth=1.5, fliersize=4)
plt.title("Box plot of total bill by day and meal time")
plt.xlabel("Day of the week")
plt.ylabel("Total Bill")
plt.show()
```

```python
sns.violinplot(x='species', y = 'petal_length', data = iris)
```

```python
corr = tips.corr()
sns.heatmap(corr)
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights, cmap='Blues', annot=True, fmt = 'd')
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passengers per month")
plt.show()
```

```python
sns.pairplot(data = iris)
plt.show()
```

```python
sns.pairplot(data = iris, hue = 'species', diag_kind='kde', palette='husl')
plt.title("Iris Dataset Pair Plot")
plt.show()
```

```python
sns.jointplot(x='sepal_length', y = 'sepal_width', data = iris)
plt.show()
```

```python
# Adding Annotations
diamonds = sns.load_dataset("diamonds")
sns.scatterplot(x = 'carat', y = 'price', data = diamonds)
plt.text(1,10000,"Large, Expensive Diamonds", fontsize = 12, color='red')
plt.text(2.5,5000,"Small, affordable Diamonds", fontsize=12, color='blue')
plt.title("Diamond Prices by Carat")
plt.xlabel("Carat")
plt.ylabel('Price')
plt.show()
```
